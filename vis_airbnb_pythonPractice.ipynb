```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
listings = pd.read_csv('Airbnb_NYC_2019.csv')
```

```python
listings
```

```python
# create a bar chart that displays a count of Airbnb listings in each neighborhood group of New York City from the listings data frame.
# use the count plot method from the seaborn library followed by the show method from the matplotlib.pyplot

sn.countplot(x = 'neighbourhood_group', data = listings)
plt.show()
```

```python
# create a bar chart that displays the average price of Airbnb listings in each neighborhoo group of NYC

sn.barplot(x = 'neighbourhood_group', y = 'price', data = listings)
plt.show()
```

```python
sn.barplot(x = 'neighbourhood_group', y = 'price', data = listings, errorbar = ('ci', False))
plt.show()
```

```python
# use the hist method from matplotlib.pyplot to create a histogram that displays a distribution of the quantitative data in the price column in listings
plt.hist(listings['price'])
plt.xlabel('price (in US dollars)')
plt.show
```

```python
#
plt.hist(listings['price'], bins = np.arange(0, 1100, 40)) # 1st - left end of the first bin, 2nd - the right end of the last bin, 3rd - step size
plt.xlabel('price (in US dollars)')
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.xlim(0, 1100)
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings['number_of_reviews'], s = 5)
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.xlim(0, 1100)
plt.show()
```
