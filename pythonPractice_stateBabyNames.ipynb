```python
df = pd.read_csv('state_baby_names.csv')
```

```python
state_babies = df
```

```python
state_babies.head()
```

```python
# What were the 5 most popular baby names in 2014 in California?
# first - slice out the rows for the years 2014 using the loc method
state_babies_2014 = state_babies.loc[state_babies['Year'] == 2014, :]

```

```python
state_babies_2014
```

```python
# Slice out the rows of the states babies 2014 data frame that correspond to the state California
# use loc method again
ca_babies_2014 = state_babies_2014.loc[state_babies_2014['State'] == 'CA', :]
```

```python
ca_babies_2014
```

```python
# sort the rows in descending order
sorted_ca_2014 = ca_babies_2014.sort_values("Count", ascending = False)
```
