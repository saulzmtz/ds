```python
# loc method allows you to access a specific groups of rows and columns from a Pandas data frame, the method receives two inputs, the first describes which rows to take from the data frame
# the second describes which columns to take from the data frame and loc returns a new df based on the rows and columns you specified
ca_babies = state_babies.loc[state_babies['State'] == 'CA', :]
```

```python
ca_babies
```

```python
# continued from comments below
# need an aggregation function that takes in a Pandas series as input and finds the most popular baby name in that series
# start with keyword def, followed by name of the function, then parenthesis and in the par type s which represents the input that this function receives, then end with a colon and enter for new lin
def popular(s): # s will be a pandas series containing baby names
# easier if function receive baby names in order of highes to lowest, so sort CA babies before calling the groupby and ag method below (see sort_values edit below)
  return s.iloc[0] # returning the item in s that has index 0  # after this call the ag method below and pass in the function popular
```

```python
# Group the CA babies by both year and gender using the group by method
# Calling the groupby method on a data frame does not return a new data frame, it returns an object called a data frame groupby object
# this 'dataframe groupby object' contains information about the groups that were created
# use another method that operates on those groups to create a new df that will be utilized instead - called the ag method
# the ag method can receive an aggregation function as input and use it to aggregate the groups of data
# ag applies the aggregation function to every column in each group - each column of a Pandas df is a Pandas series
# in this example the df has been grouped by year and gender -- need an aggregation function that takes in a Pandas series as input and finds the most popular baby name in that series
# start with creating a cell above
ca_grouped = ca_babies.sort_values('Count', ascending = False).groupby(['Year', 'Gender']).agg(popular)  # call the agg method and pass in function popular created above and assign all this to new var
```
